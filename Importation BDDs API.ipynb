```python
import requests
import json
import pandas as pd
```

## Notebook d'obtention des bases de données avec utilisation d'un API

Ci-dessous, le code de requête à l'API data.gouv pour l'obtention des données specialites.csv et ips.csv. En raison de la longueur de compilation du code, cette partie a été séparée du reste du notebook.

L'API data.gouv limitant chaque requête à 100 lignes maximum, une succession de requêtes sont envoyées jusqu'à l'obtention de la base complète.

### Obtention de la bdd specialites.csv

```python
response = requests.get("https://data.education.gouv.fr/api/explore/v2.1/catalog/datasets/fr-en-effectifs-specialites-triplettes-1ere-generale/records?limit=100")
rep = response.json()
df = pd.DataFrame(rep['results'])
for i in range(1, (rep['total_count']//100)+1) :
    response = requests.get("https://data.education.gouv.fr/api/explore/v2.1/catalog/datasets/fr-en-effectifs-specialites-triplettes-1ere-generale/records?limit=100&offset=" + str(i))
    rep = response.json()
    df2 = pd.DataFrame(rep['results'])
    df = pd.concat([df, df2], ignore_index = True)
specialites = df
```

```python
specialites
```

```python
specialites.to_csv("specialites.csv", index = False)
```

### Obtention de la bdd ips.csv

```python
response = requests.get("https://data.education.gouv.fr/api/explore/v2.1/catalog/datasets/fr-en-ips-lycees-ap2022/records?limit=100")
rep = response.json()
df = pd.DataFrame(rep['results'])
for i in range(1, (rep['total_count']//100)+1) :
    response = requests.get("https://data.education.gouv.fr/api/explore/v2.1/catalog/datasets/fr-en-ips-lycees-ap2022/records?limit=100&offset=" + str(i))
    rep = response.json()
    df2 = pd.DataFrame(rep['results'])
    df = pd.concat([df, df2], ignore_index = True)
ips = df
```

```python
ips.fillna("None")
ips
```

```python
ips.to_csv("ips.csv", index = False)
```
